```python
qlik_bestand = pd.read_csv(qlik_bestand_csv, encoding="utf-8-sig", low_memory=False)
qlik_bestand = qlik_bestand.rename(
    columns={
        k: v
        for k, v in {"Lager": "WERKS", "Btrb": "WERKS", "Artikel (Nr)": "MATNR"}.items()
        if k in qlik_bestand.columns
    }
)
_werks_num = pd.to_numeric(qlik_bestand["WERKS"], errors="coerce")
qlik_bestand["WERKS"] = _werks_num.map(
    lambda x: f"{int(x):04d}" if pd.notna(x) else pd.NA
)
qlik_bestand["MATNR"] = qlik_bestand["MATNR"].astype(str)

def _xyz_from_buz(buz):
    if pd.isna(buz):
        return pd.NA
    s = str(buz).strip()
    return s[1] if len(s) >= 2 else pd.NA


qlik_bestand["XYZ"] = qlik_bestand["BUZ-Klasse"].map(_xyz_from_buz)
qlik_bestand.head()
```

```python
# Per MATNR: WERKS counts for XYZ (2nd char of BUZ-Klasse) buckets; exclude ND dispositions
_not_nd = ~(
    qlik_bestand["Dispositionsmerkmal"]
    .astype(str)
    .str.strip()
    .str.upper()
    .eq("ND")
)
_q = qlik_bestand.loc[_not_nd].copy()

matnr_xyz_werks_counts = (
    _q.assign(
        is_x=_q["XYZ"].isin(["X", "0"]),
        is_y=_q["XYZ"].eq("Y"),
        is_z=_q["XYZ"].eq("Z"),
    )
    .groupby("MATNR", as_index=False)
    .agg(
        werks_xyz_X=("is_x", "sum"),
        werks_xyz_Y=("is_y", "sum"),
        werks_xyz_Z=("is_z", "sum"),
    )
)
matnr_xyz_werks_counts.head()
```

```python
sortiment_file = full_extract_path / "Sortiment_MM_Rundungswerte 2026-02-03.xlsx"

sortiment = pd.read_excel(
    sortiment_file,
    dtype={"MATNR": str},   # ensures MATNR is loaded as object/string
    sheet_name="VORSCHLAGSWERTE"  # specify if not the first sheet
)

sortiment = sortiment.groupby(["MATNR"]).agg({"AT_SORTIMENT_EBENE1":"first", "AT_SORTIMENT_EBENE2":"first", "AT_SORTIMENTSBAUSTEIN":"first", "Rolle":"first"})
sortiment.reset_index(inplace=True)

sortiment.head()
```

```python
# Join region (WERKS + MATNR) → qlik_bestand → artstamm (3 cols) → mara (MATKL=Warengruppe) → sortiment
_r = region.copy()
_r["WERKS"] = pd.to_numeric(_r["WERKS"], errors="coerce").map(
    lambda x: f"{int(x):04d}" if pd.notna(x) else pd.NA
)
_r["MATNR"] = _r["MATNR"].astype(str)

_qlik = qlik_bestand.drop_duplicates(subset=["MATNR", "WERKS"], keep="first")
_art = artstamm[["MATNR", "Artikelkurztext1", "Artikelkurztext2"]].drop_duplicates(
    subset=["MATNR"], keep="first"
)
_mara = (
    mara[["MATNR", "MATKL"]]
    .drop_duplicates(subset=["MATNR"], keep="first")
    .assign(MATNR=lambda d: d["MATNR"].astype(str), MATKL=lambda d: d["MATKL"].astype(str))
    .rename(columns={"MATKL": "Warengruppe_mara"})
)
_sort = sortiment.copy()
_sort["MATNR"] = _sort["MATNR"].astype(str)

region_merged = (
    _r.merge(_qlik, on=["MATNR", "WERKS"], how="left", suffixes=("", "_qlik"))
    .merge(_art, on="MATNR", how="left")
    .merge(_mara, on="MATNR", how="left")
    .merge(_sort, on="MATNR", how="left", suffixes=("", "_sortiment"))
)
region_merged["Kapazitätshöchstbestand - Anpassung"] = ""
region_merged["Rundungswert - Anpassung"] = ""
region_merged["Dispomerkmal - Anpassung"] = ""
_xyz_cnt = matnr_xyz_werks_counts.copy()
_xyz_cnt["MATNR"] = _xyz_cnt["MATNR"].astype(str)
region_merged = region_merged.merge(_xyz_cnt, on="MATNR", how="left")

_max_b = pd.to_numeric(region_merged["Max. Bestand"], errors="coerce")
_kap_max = pd.to_numeric(region_merged["Kapazitätshöchstbestand"], errors="coerce")
_bs_col = (
    "Bestandsmenge"
    if "Bestandsmenge" in region_merged.columns
    else "Bestandsmng. in LME"
)
_best = pd.to_numeric(region_merged[_bs_col], errors="coerce")
_hinweis = ((_max_b > 0.9 * _kap_max) | (_best > _kap_max)) & (_kap_max >0)
region_merged["Hinweis Kapazitätsgrenze"] = np.where(
    _hinweis, "Überprüfung Kapazitätsgrenze", ""
)

_sort_hinweis = (
    (region_merged["BUZ-Klasse"].isin(["CY", "CZ"])) &
    (region_merged["AT_SORTIMENTSBAUSTEIN"].fillna("") != "Kernsortiment BGW")
    #(region_merged["Kernsortiment"].fillna("") != "Kernsortiment BGW")
)

region_merged["Hinweis Sortiment"] = np.where(
    _sort_hinweis, "Überprüfung ob ND gesetzt werden sollte", ""
)

_inv_lme = pd.to_numeric(region_merged["Bestandsmng. in LME"], errors="coerce").fillna(0)
_gw = pd.to_numeric(region_merged["Gesamtwert"], errors="coerce")
_gld_preis = pd.to_numeric(region_merged["GLD-Preis"], errors="coerce")
region_merged["cost_used"] = np.where(
    _inv_lme != 0,
    _gw / _inv_lme,
    _gld_preis,
)
region_merged.head()
```

```python
def autofit_columns(ws, double_wide_cols=None):
    for col in ws.columns:
        max_length = max((len(str(cell.value)) for cell in col if cell.value is not None), default=0)
        width = max_length + 4
        col_letter = col[0].column_letter
        if double_wide_cols and col[0].value in double_wide_cols:
            width *= 2
        ws.column_dimensions[col_letter].width = width
```

```python
def build_sortiment_sheet(writer, df_nl):
    df_sort = df_nl[df_nl["Hinweis Sortiment"] != ""].copy()
    df_sort = df_sort[
        pd.to_numeric(df_sort["Mindestmeldebestand"], errors="coerce")
        > pd.to_numeric(df_sort["Meldebestand_alt"], errors="coerce")
    ].copy()
    df_sort["Artikelkurztext"] = (
        df_sort["Artikelkurztext 1"].fillna("") + " " + df_sort["Artikelkurztext 2"].fillna("")
    ).str.strip()
    cols_base = [
        "Niederlassung", "Artikelnummer", "Artikelkurztext", "Sparte", "BUZ-Klasse", "Kernsortiment",
        "Hinweis Sortiment", "Mindestmeldebestand", "Dispomerkmal",
        "Sonderbeschaffungsschlüssel", "Lagerhaltzungskz.",
        "Bestandsmenge", "Externe Zugriffe", "Jahresabsatz LME",
    ]
    df_export = df_sort.reindex(columns=[c for c in cols_base if c in df_sort.columns])
    hinweis_pos = df_export.columns.get_loc("Hinweis Sortiment") + 1
    df_export.insert(hinweis_pos, "Aktivität", "")
    df_export.insert(hinweis_pos + 1, "Kommentar", "")
    df_export.to_excel(writer, index=False, sheet_name="CY-CZ Artikel")
    ws = writer.sheets["CY-CZ Artikel"]
    light_green = PatternFill(start_color="C6EFCE", end_color="C6EFCE", fill_type="solid")
    from openpyxl.utils import get_column_letter
    from openpyxl.worksheet.datavalidation import DataValidation
    aktivitaet_col_idx = df_export.columns.get_loc("Aktivität") + 1
    kommentar_col_idx = df_export.columns.get_loc("Kommentar") + 1

    for col_idx in [aktivitaet_col_idx, kommentar_col_idx]:
        for row in ws.iter_rows(min_row=2, max_row=ws.max_row, min_col=col_idx, max_col=col_idx):
            for cell in row:
                cell.fill = light_green

    # Avoid invalid ranges like A2:A1 when only header exists.
    if ws.max_row >= 2:
        dv = DataValidation(
            type="list",
            formula1='"Artikel soll weiter im Bestand bleiben,Wir werden Artikel auf ND setzen"',
            allow_blank=True,
            showDropDown=False,
        )
        ws.add_data_validation(dv)
        act_letter = get_column_letter(aktivitaet_col_idx)
        dv.sqref = f"{act_letter}2:{act_letter}{ws.max_row}"

    ws.auto_filter.ref = ws.dimensions
    autofit_columns(ws, double_wide_cols=["Aktivität", "Kommentar"])
```

```python
def build_kapazitaet_sheet(writer, df_nl):
    df_kap = df_nl[df_nl["Hinweis Kapazitätsgrenze"] != ""].copy()
    df_kap["Artikelkurztext"] = (
        df_kap["Artikelkurztext 1"].fillna("") + " " + df_kap["Artikelkurztext 2"].fillna("")
    ).str.strip()

    cols = [
        "Niederlassung", "Artikelnummer", "Artikelkurztext", "Sparte", "BUZ-Klasse",
        "AT_SORTIMENTSBAUSTEIN", "Bestandsmenge", "Rundungswert",
        "Kapazitätshöchstbestand", "Kapazitätshöchstbestand - Anpassung",
    ]
    df_export = df_kap.reindex(columns=[c for c in cols if c in df_kap.columns])
    df_export = df_export.rename(columns={"AT_SORTIMENTSBAUSTEIN": "Kernsortiment"})

    df_export.to_excel(writer, index=False, sheet_name="Kapazitätsgrenze")
    ws = writer.sheets["Kapazitätsgrenze"]

    light_green = PatternFill(start_color="C6EFCE", end_color="C6EFCE", fill_type="solid")
    kap_col_idx = df_export.columns.get_loc("Kapazitätshöchstbestand - Anpassung") + 1
    for row in ws.iter_rows(min_row=1, max_row=ws.max_row, min_col=kap_col_idx, max_col=kap_col_idx):
        for cell in row:
            cell.fill = light_green

    ws.auto_filter.ref = ws.dimensions
    autofit_columns(ws)
```

## Determine impact on purchasing

#### Logic description

a) Filter on any items with Dispositionsmerkmal != ND \
b) Determine the respective 90% quantile \
c) Compare Meldebestand with availability_limit and make it a binary 1 "Meldebestand erhöht" if Meldebestand < availability_limit \
d) In those cases where Meldebestand erhöhte calculate the delta, else fill it with 0 \
e) Then group by MATNR and count total increase and sum over Meldebestand_erhöht \
f) After that merge with the dataframe containing number of x, y and z locations \

```python
def _availability_limit(row):
    if USE_GLOBAL_LIMIT:
        return row["q90_alle_NL"]
    if pd.isna(row["separate_betrachtung"]) or row["separate_betrachtung"] == 0:
        return row["q90_alle_NL"]
    # separate_betrachtung == 1
    excluded = row["excluded_top_werks"]
    excluded_set = set(str(excluded).split(",")) if pd.notna(excluded) else set()
    if row["WERKS"] not in excluded_set:
        return row["q90_ohne_top3"]
    return row["q90_alle_NL"]
```

```python
USE_GLOBAL_LIMIT = False

# a) Filter qlik_bestand: exclude ND Dispositionsmerkmal
_not_nd_mask = (
    qlik_bestand["Dispositionsmerkmal"]
    .astype(str)
    .str.strip()
    .str.upper()
    .ne("ND")
)
melde_df = qlik_bestand.loc[
    _not_nd_mask, ["MATNR", "WERKS", "Meldebestand", "Bestandswert (in €)", "Jahresabsatz LME", "Max. Bestand"]
].copy()
melde_df["MATNR"] = melde_df["MATNR"].astype(str)

# b) Merge with stats_quantile
_sq = stats_quantile[
    ["MATNR", "q90_alle_NL", "q90_ohne_top3", "excluded_top_werks", "separate_betrachtung"]
].copy()
_sq["MATNR"] = _sq["MATNR"].astype(str)
melde_df = melde_df.merge(_sq, on="MATNR", how="left")

# c) Determine availability_limit per row
melde_df["availability_limit"] = melde_df.apply(_availability_limit, axis=1)

# d) Binary flag: 1 if Meldebestand < availability_limit
melde_df["Meldebestand"] = pd.to_numeric(melde_df["Meldebestand"], errors="coerce")
melde_df["Meldebestand_erhöht"] = (
    melde_df["Meldebestand"] < melde_df["availability_limit"]
).astype(int)

# e) Delta where erhöht == 1, else 0
melde_df["Meldebestand_delta"] = (
    (melde_df["availability_limit"] - melde_df["Meldebestand"])
    .where(melde_df["Meldebestand_erhöht"] == 1, other=0)
)
melde_df["Jahresabsatz LME"] = pd.to_numeric(melde_df["Jahresabsatz LME"], errors="coerce")
melde_df["Jahresabsatz_erhöht"] = melde_df["Jahresabsatz LME"].where(
    melde_df["Meldebestand_erhöht"] == 1, other=0
)

# Max. Bestand: if Meldebestand > Max. Bestand, use Meldebestand * 2, else keep Max. Bestand
melde_df["Max. Bestand"] = pd.to_numeric(melde_df["Max. Bestand"], errors="coerce")
melde_df["Max. Bestand_adj"] = melde_df["Max. Bestand"].where(
    melde_df["Meldebestand"] <= melde_df["Max. Bestand"],
    other=melde_df["Meldebestand"] * 2,
)

# Bestellmenge = Max. Bestand (adjusted) - Meldebestand
melde_df["Bestellmenge"] = (
    (melde_df["Max. Bestand_adj"] - melde_df["Meldebestand"])
    .where(melde_df["Meldebestand_erhöht"] == 1, other=0)
)

# f) Group by MATNR: count locations needing increase + sum of delta
melde_summary = (
    melde_df.groupby("MATNR", as_index=False)
    .agg(
        Meldebestand_erhöht_count=("Meldebestand_erhöht", "sum"),
        Meldebestand_delta_sum=("Meldebestand_delta", "sum"),
        Bestandswert_sum=("Bestandswert (in €)", "sum"),
        Jahresmenge_Meldebestand_erhöht=("Jahresabsatz_erhöht", "sum"),
        Jahresmenge_gesamt=("Jahresabsatz LME", "sum"),
        Bestellmenge_sum=("Bestellmenge", "sum"),
    )
)

melde_summary["Bestellmenge_sinnvoll"] = (
    melde_summary["Bestellmenge_sum"] < melde_summary["Jahresmenge_Meldebestand_erhöht"]
).astype(int)

# g) Merge with X/Y/Z location counts per MATNR
_xyz = matnr_xyz_werks_counts.copy()
_xyz["MATNR"] = _xyz["MATNR"].astype(str)
melde_summary = melde_summary.merge(_xyz, on="MATNR", how="left")

_artikel_info = (
    qlik_bestand[["MATNR", "Sparte", "Artikel"]]
    .drop_duplicates(subset="MATNR", keep="first")
    .assign(MATNR=lambda d: d["MATNR"].astype(str))
)
melde_summary = melde_summary.merge(_artikel_info, on="MATNR", how="left")

melde_summary.sort_values("Meldebestand_erhöht_count", ascending=False).head()
```
